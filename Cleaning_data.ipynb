```python
import pandas as pd
```

```python
df = pd.read_csv("london.csv")
```

```python
df = df[df['duration'].notna()]
print(len(df))
df = df[df['end_station_id'].notna()]
print(len(df))
df = df[df['start_station_name'].notna()]
print(len(df))
df = df[df['end_rental_date_time'].notna()]
print(len(df))
```

```python
df3 = df2.rename(columns={'station_id': 'start_station_id'})
df3.drop(columns=['station_name'],inplace=True)
```

```python
df4 = pd.merge(df, df3, on=['start_station_id'])
```
